```python
import pandas as pd
import numpy as np
```

```python
archivo = "data.xlsx" 
datos = pd.read_excel(archivo, header=[6], sheet_name='Potosí')
datos = datos.drop('Unnamed: 0', axis=1)
datos = datos.drop(list(range(69, 76)))
datos
```

```python
data = pd.concat( [datos.iloc[:,0] , datos.iloc[:,31:] ] , axis = 1 )
data
```

```python
nan_index = data[data.isnull().all(axis=1)].index.tolist()
dfs = {}

for i in range(0,len(nan_index)-1):
    dataframe = data.iloc[nan_index[i]+1:nan_index[i+1],:].reset_index(drop=True)
    dfs[dataframe.iloc[0,0]] = dataframe.iloc[1:, :]

dataframe = data.iloc[nan_index[-1]+1:,:].reset_index(drop=True)
dfs[dataframe.iloc[0,0]] = dataframe.iloc[1:, :]
```

```python
#1
for prod in list(dfs.keys()):
  pos_max_value = dfs[prod].iloc[:,-1].idxmax()
  print(prod, " => ", dfs[prod].iloc[pos_max_value-1,0], ": ", dfs[prod].iloc[pos_max_value-1, -1])
```

```python
#2
keys = list(dfs.keys())
max_crec_prod = keys[0]
max_value = np.sum(dfs[keys[0]], axis=0).reset_index().iloc[1:, 1:].sum()[0]

for i in range(1, len(keys)):
  value = np.sum(dfs[keys[i]], axis=0).reset_index().iloc[1:, 1:].sum()[0]
  if value > max_value:
    max_value = value  
    max_crec_prod = keys[i]
print(max_crec_prod, " : ", max_value)
```
