```python
excel = pd.read_excel('censoAGR_SCZ.xls', header=[9]) #lectura del archivo excel

# excel.reset_index(inplace=True)
excel = excel.drop('Unnamed: 0', axis=1) # elimnar la primera columnna vacia 
excel = excel.drop(excel.iloc[:, 40:], axis =1) # eliminar las ultimas columnas
excel.drop(list(range(56, 59)), inplace=True) # elminar los ultimos registros vacios
excel.head()
```

```python
#primera pregunta
pp = excel[["Código", "Municipio", "Si", "No"]] #extraer los datos para la primera pregunta
# pp['Código'].apply(pd.to_numeric)
pp.head()
```

```python
#segunda pregunta
sp = excel[["Código", "Municipio", "Si.1", "No.1", "Sin Especificar"]] # extraer para la segunda pregunta
sp = sp.rename(columns={'Si.1': "Si", 'No.1': "No"})
sp.head()
```

```python
#tercera pregunta
tp = excel[["Código", "Municipio", "Si.2", "No.2", "Sin Especificar.1"]]
tp = sp.rename(columns={'Si.1': "Si", 'No.1': "No"})
tp.head()
```

```python
#codigo para poder automatizar el proceso de separacion de los datos
dfms = [f'p{i}' for i in range(2, 14)] #indices para dataframes
dfms_pandas = {}
# left = excel.iloc[:,0:2]# ["Código", "Municipio"]
num_col = 4
for df in dfms:
  dfms_pandas[df] = pd.DataFrame({
        "Código": excel.iloc[:, 0], 
        "Municipio":excel.iloc[:, 1],
        "Si":excel.iloc[:, num_col], 
        "No":excel.iloc[:, num_col+1], 
        "Sin especificar":excel.iloc[:, num_col+2]
      })
  num_col +=3 
```
