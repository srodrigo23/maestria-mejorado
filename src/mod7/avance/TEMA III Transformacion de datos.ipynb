# Datos categoricos

Cuando colectamos datos estos pueden ser clasificados en dos principales categorias:
1. Datos Categoricos
2. Datos numéricos

Una variable categórica solo acepta una cantidad limitada de valores.

Por ejemplo en una encuesta se pregunta con qué frecuencia desayuna y tenemos cuatro opciones: "Nunca", "Rara vez", "La mayoría de los días" o "Todos los días". 

En este caso, los datos son categóricos, **porque las respuestas se clasifican en un conjunto fijo de categorías**.

Los datos categoricos pueden ser clasidficados en:

1. Datos nominales
Los datos nominales son un tipo de datos que consisten en categorías que no se pueden ordenar ni clasificar. También se denominan escalas nominales. Los datos nominales no se pueden clasificar ni medir de ninguna manera, pero a veces pueden ser tanto cualitativos como cuantitativos. Algunos ejemplos de datos nominales son los símbolos, las palabras, las letras y el género de una persona.

2. Datos ordinales
Los datos ordinales son una categoría de datos que tiene un orden natural. Se utilizan a menudo en encuestas, cuestionarios y campos como las finanzas y la economía. Lo que diferencia a los datos ordinales es que es imposible determinar la diferencia entre los valores de los datos. Las tallas de ropa son un ejemplo (pequeño, mediano y grande no son diferencias medibles, pero están claramente ordenadas para mostrar comparaciones de tallas).

# Visualizacion

Para representar visualmente datos categóricos y detectar tendencias, diferencias y distribuciones de categorías, se utilizan comúnmente los siguientes gráficos:

1. **Gráficos de barras:** a menudo utilizados tanto para datos nominales como ordinales, los gráficos de barras muestran la frecuencia de las categorías con barras rectangulares.
2. **Gráficos circulares:** útiles para datos nominales para mostrar la proporción de categorías como porciones de un gráfico circular (aunque menos favorecidos en análisis complejos). Puede ver un ejemplo claro de este tipo de análisis en el segmento “Ejemplos de datos categóricos”.
3. **Gráficos de barras apiladas:** compare la frecuencia de categorías en diferentes grupos.

# Test estadísticos para data Categórica.

Aunque los datos categóricos suelen ser no numéricos, existen pruebas estadísticas específicas que pueden analizar estos datos desde una perspectiva estadística y revelar relaciones entre variables categóricas. Las pruebas más populares para variables categóricas son:

1. **Prueba de independencia de chi-cuadrado:** la prueba de chi-cuadrado determina si existe una asociación significativa entre dos variables categóricas.
2. **Prueba exacta de Fisher:** se utiliza cuando el tamaño de la muestra es pequeño. Al igual que la prueba de chi-cuadrado, verifica la asociación entre variables categóricas.
3. **Bondad de ajuste de chi-cuadrado:** prueba si la distribución observada de una variable categórica coincide con una distribución esperada.
4. **Regresión logística:** analiza las relaciones entre una o más variables categóricas (variables independientes) y un resultado categórico binario (variable dependiente). También se puede extender a resultados multinomiales (regresión logística multinomial).

# ¿Por qué?

Algunos algoritmos de aprendizaje automático pueden trabajar directamente con datos categóricos según la implementación, como un árbol de decisiones, pero la mayoría requiere que las variables de entrada o salida sean un número o un valor numérico. Esto significa que todos los datos categóricos deben asignarse a números enteros.

# Manejando variables categóricas

**1. Original Encoding**

Es un método de conversión de datos para prepararlos para un algoritmo. **La codificación ordinal** asigna cada valor único a un entero diferente.

No todas las variables categóricas tienen un orden claro en sus valores, pero las que sí lo tienen se denominan variables ordinales. En el caso de los **modelos basados en árboles (como los árboles de decisión y los bosques aleatorios)**, se puede esperar que la codificación ordinal funcione bien con las variables ordinales.

**2. One-Hot Encoding**

En **la codificación one-hot** convertimos cada valor categórico en una nueva columna categórica y asignamos un valor binario de 1 o 0 a esas columnas. Cada valor entero se representa como un vector binario. Todos los valores son cero y el índice se marca con un 1.

A diferencia de la codificación ordinal, la codificación one-hot no presupone un ordenamiento de las categorías. Por lo tanto, se puede esperar que este enfoque funcione particularmente bien si no existe un ordenamiento claro en los datos categóricos (por ejemplo, "Rojo" no es ni más ni menos que "Amarillo"). Nos referimos a las variables categóricas sin una clasificación intrínseca como variables nominales.

La codificación one-hot generalmente no funciona bien si la variable categórica adopta una gran cantidad de valores (es decir, generalmente no se la utilizará para variables que toman **más de 15 valores diferentes**).

# Usando sklearn

Scikit-learn, conocida como sklearn is una libreria abierta de machine learning y modelación de datos para python. Proporciona un conjunto completo de algoritmos de aprendizaje supervisados y no supervisados, que cubren áreas como:

1. **Clasificación:** Identificación de la categoría a la que pertenece un objeto.
2. **Regresión**: Predicción de un atributo de valor continuo asociado con un objeto.
3. **Agrupamiento:** Agrupamiento automático de objetos similares en conjuntos, con modelos como k-means.
4. **Reducción de la dimensionalidad:** Reducción de la cantidad de atributos en los datos para su resumen, visualización y selección de características, con modelos como el Análisis de componentes principales (PCA).
5. **Selección de modelos:** Comparación, validación y elección de parámetros y modelos.
6. **Preprocesamiento:** Extracción y normalización de características, incluida la definición de atributos en datos de imágenes y texto.

Scikit-learn está escrito principalmente en Python y utiliza NumPy ampliamente para operaciones de álgebra lineal y matriz de alto rendimiento, no olvidar que esta libreria **no puede recibir datos faltantes**.

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

# Crear el dataset
data = pd.DataFrame({
    'Nombre': ['Ana', 'Juan', 'Luis', 'María'],
    'Ciudad': ['Nueva York', 'Los Ángeles', 'Nueva York', 'Chicago'],
    'Rango de Edad': ['Joven', 'Adulto', 'Adulto', 'Joven']
})

# Mostrar el dataset original
print("Dataset original:")
print(data)

# =============================
# 1. Aplicar OneHotEncoder a 'Ciudad'
# =============================
# Crear el codificador
one_hot_encoder = OneHotEncoder(sparse=False)  # sparse=False para obtener una matriz densa

# Codificar la columna 'Ciudad'
ciudad_encoded = one_hot_encoder.fit_transform(data[['Ciudad']])

# Convertir a DataFrame
ciudad_encoded_df = pd.DataFrame(ciudad_encoded, columns=one_hot_encoder.get_feature_names_out(['Ciudad']))

# =============================
# 2. Aplicar OrdinalEncoder a 'Rango de Edad'
# =============================
# Crear el codificador
ordinal_encoder = OrdinalEncoder(categories=[['Joven', 'Adulto', 'Senior']])  # Especificar el orden de las categorías

# Codificar la columna 'Rango de Edad'
rango_edad_encoded = ordinal_encoder.fit_transform(data[['Rango de Edad']])

# Agregar la columna codificada al DataFrame original
data['Rango de Edad Codificado'] = rango_edad_encoded

# =============================
# Combinar el dataset final
# =============================
# Concatenar las columnas codificadas (OneHotEncoder) con el dataset original
resultado = pd.concat([data, ciudad_encoded_df], axis=1)

# Mostrar el dataset final
print("\nDataset transformado:")
print(resultado)
```

# **Estandarización y normalización**

Inicialmente realizamos un análisis exploratorio para comprender la naturaleza de los datos, luego de ser necesario se aplica alguna estrategia de limpieza de datos. En la siguiente etapa se realiza el preprocesamiento de datos, datos que generalmente son usados para aplicar en una etapa posterior modelado.

En el preprocesamiento de datos se utilizan con frecuencia dos técnicas: **normalización o estandarización.**
Se aplicará una u otra, normalización o estandarización, en función de los datos que se tengan a mano, del algoritmo que se utilice en la fase de modelado, de la escala y del formato de los datos


Por tanto, la elección entre una técnica u otra empieza por entender los datos. Por eso es tan importante el análisis exploratorio.

**Normalización**

La normalización se utiliza para redimensionar los valores de los datos a un rango común sin distorsionar las diferencias en los rangos de valores.
Hay una diferencia entre datos e información. La normalización se utiliza para modificar el formato de los datos, los cuales tomarán valores en un rango de $[-1, 1]$ o $[0,1]$, pero no puede modificar la información.

Se permite modificar los datos. Para ello tenemos varias estrategias, varias técnicas matemáticas. La normalización es matemática

Denominado método **escalado mínimo-máximo** Esto es útil cuando los datos tienen diferentes escalas y se quieren comparar de forma equitativa.
Por ejemplo, imaginemos que tengo dos variables en mi conjunto de datos: edad y salario. Y estoy creando un modelo para predecir si un empleado renunciará o no.
Entonces, tengo dos variables de entrada, edad y salario. La edad suele ser de dos dígitos, ¿verdad? Si es un empleado, probablemente tenga más de 10 años o más. Incluso puede tener más de 100 años, así que también consideraré tres dígitos, sin problema. Pero normalmente, la escala es de dos dígitos, ¿verdad? Edad, 30 años, 45, 55, 18 años, etcétera.
El salario probablemente esté en los miles. Entonces, la persona gana 8 mil, 10 mil, 15 mil, 30 mil, 50 mil, etcétera. ¿Ok? Las variables están en diferentes escalas.

**Estandarización**

La estandarización se utiliza para cambiar el tamaño de los datos de modo que tengan una media de 0 y una desviación estándar de 1.

Esto transforma los datos en una distribución normal estándar. De ahí el nombre de estandarización.



Esto es útil en algoritmos que asumen que todas las características estan centrada alrededor de 0 y tienen la misma variación.


Aplicamos normalización o estandarización dependiendo de los datos y del algoritmo que se utilice.

**Aplicabilidad:** La normalización es más útil cuando los datos no siguen una distribución normal y necesitas una escala común. La estandarización es preferible cuando los datos siguen una distribución normal.

**Elección:** La elección entre normalización y estandarización depende de tu conjunto de datos y del algoritmo que estés utilizando. Algunos algoritmos, como las redes neuronales, suelen beneficiarse de la normalización, mientras que otros, como los algoritmos basados en la distancia, pueden preferir la estandarización.

¿Cuántos algoritmos tenemos hoy? Cientos. Considerando el algoritmo estándar y sus variaciones, hay cientos. En otras palabras, para cada algoritmo a usar **hay que revisar la documentación**, las especificaciones del algoritmo, buscar la documentación, por ejemplo, en el framework, para saber si ese algoritmo espera recibir datos con una distribución normal, una distribución normal estándar, datos en la misma escala, para poder decidir entre normalización y estandarización.

En la práctica, los algoritmos de aprendizaje no supervisado se benefician más de la estandarización que de la normalización

# Ejemplos:

Min-Max Scaler: Se tiene un dataset con información sobre el rendimiento de estudiantes en tres exámenes. Las puntuaciones están en diferentes rangos:

1. Examen 1: Rango de 0 a 100.
2. Examen 2: Rango de 0 a 50.
3. Examen 3: Rango de 0 a 200.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Dataset de ejemplo
data = pd.DataFrame({
    'Examen 1': [50, 80, 90, 60],
    'Examen 2': [30, 25, 45, 40],
    'Examen 3': [100, 150, 180, 120]
})

# Mostrar el dataset original
print("Dataset original:")
print(data)

# Crear el MinMaxScaler
scaler = MinMaxScaler()

# Escalar los datos
data_scaled = scaler.fit_transform(data)

# Convertir el resultado a un DataFrame
data_scaled_df = pd.DataFrame(data_scaled, columns=data.columns)

# Mostrar los datos escalados
print("\nDataset escalado (MinMaxScaler):")
print(data_scaled_df)
```

Tienes un dataset con los ingresos anuales (en miles de dólares) y el puntaje de crédito de un grupo de personas. Las columnas están en diferentes escalas, pero quieres estandarizarlas usando el Z-score, de modo que cada columna tenga una media de 0 y desviación estándar de 1.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Dataset de ejemplo
data = pd.DataFrame({
    'Ingresos': [40, 50, 60, 70],
    'Puntaje de Crédito': [600, 700, 800, 750]
})

# Mostrar el dataset original
print("Dataset original:")
print(data)

# Crear el StandardScaler
scaler = StandardScaler()

# Escalar los datos
data_scaled = scaler.fit_transform(data)

# Convertir el resultado a un DataFrame
data_scaled_df = pd.DataFrame(data_scaled, columns=data.columns)

# Mostrar los datos estandarizados (Z-score)
print("\nDataset estandarizado (Z-score):")
print(data_scaled_df)
```

# Principal Component Analysis PCA

El análisis de componentes principales (PCA) es un método de reducción de dimensionalidad y aprendizaje automático que se utiliza para simplificar un conjunto grande de datos en un conjunto más pequeño, manteniendo al mismo tiempo patrones y tendencias significativos.

Reducir la cantidad de variables de un conjunto de datos naturalmente implica sacrificar la precisión, pero el truco en la reducción de la dimensionalidad es sacrificar un poco de precisión por simplicidad, ya que los conjuntos de datos más pequeños son más fáciles de explorar y visualizar, y por lo tanto hacen que el análisis de puntos de datos sea mucho más fácil y rápido para los algoritmos de aprendizaje automático sin variables extrañas que procesar.

**¿Que son los componentes principales?**

Los componentes principales son nuevas variables que se construyen como combinaciones lineales o mezclas de las variables iniciales. Estas combinaciones se realizan de tal manera que las nuevas variables (es decir, los componentes principales) no están correlacionadas y la mayor parte de la información dentro de las variables iniciales se comprime o se aprieta en los primeros componentes. Por lo tanto, la idea es que los datos de 10 dimensiones le brinden 10 componentes principales, pero el PCA intenta colocar la máxima información posible en el primer componente, luego la máxima información restante en el segundo y así sucesivamente, hasta obtener algo como lo que se muestra en el diagrama de pantalla a continuación.


Organizar la información en componentes principales de esta manera le permitirá reducir la dimensionalidad sin perder mucha información, y esto al descartar los componentes con poca información y considerar los componentes restantes como sus nuevas variables.

Geométricamente hablando, los componentes principales representan las direcciones de los datos que explican una cantidad máxima de varianza, es decir, las líneas que capturan la mayor parte de la información de los datos. La relación entre la varianza y la información aquí es que, cuanto mayor sea la varianza que lleva una línea, mayor será la dispersión de los puntos de datos a lo largo de ella, y cuanto mayor sea la dispersión a lo largo de una línea, más información tendrá. Para poner todo esto en términos simples, piense en los componentes principales como nuevos ejes que brindan el mejor ángulo para ver y evaluar los datos, de modo que las diferencias entre las observaciones sean más visibles.

Asumiendo un scatter plot con puntos como de la imagen la principal componente es la recta que coindice con lo las lineas purpuras, dado que esta recta maximiza la varianza.

El segundo componente principal se calcula de la misma manera, con la condición de que no esté correlacionado con (es decir, sea perpendicular a) el primer componente principal y que represente la siguiente varianza más alta.

A grandes rasgos los pasos que se siguen para una transformacion PCA son:

1. **Estandarización.** Cada variable debe contribuir de manera equitativa al análisis
2. **Computación de la matríz de covarianza.** En este paso el objetivo de este paso es comprender cómo las variables del conjunto de datos de entrada varían con respecto a la media entre sí, o en otras palabras, ver si existe alguna relación entre ellas y la varianza.
3. **Computación de los vectores propios y los valores propios de la matriz de covarianza** para identificar los componentes principales, los vectores propios de la matriz de covarianza son en realidad las direcciones de los ejes donde hay más varianza (más información) y que llamamos Componentes Principales.
4. **Creación de un vector característico.** En el paso anterio se realizó el cálculo de los vectores propios y su ordenación por sus valores propios en orden descendente, esto nos permite encontrar los componentes principales en orden de significación. En este paso, lo que hacemos es elegir si mantenemos todos estos componentes o descartamos los de menor significación (de valores propios bajos), y formamos con los restantes una matriz de vectores que llamamos Vector de características.
5. **Reestructuración de los datos según los ejes de componentes principales.** En este paso, que es el último, el objetivo es utilizar el vector de características formado utilizando los vectores propios de la matriz de covarianza, para reorientar los datos desde los ejes originales hacia los representados por los componentes principales (de ahí el nombre de Análisis de componentes principales). Esto se puede hacer multiplicando la transpuesta del conjunto de datos original por la transpuesta del vector de características.

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Crear un DataFrame de ejemplo con 3 variables
np.random.seed(42)
data = pd.DataFrame({
    'x1': np.random.normal(0, 1, 100),
    'x2': np.random.normal(1, 2, 100),
    'x3': np.random.normal(2, 3, 100)
})

# Visualización inicial
print("Datos originales:")
print(data.head())
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Crear la figura y los ejes 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Graficar los puntos en 3D
ax.scatter(data['x1'], data['x2'], data['x3'], c='b', marker='o', alpha=0.7, label='Puntos')

# Etiquetas de los ejes
ax.set_xlabel('X1 (Eje 1)')
ax.set_ylabel('X2 (Eje 2)')
ax.set_zlabel('X3 (Eje 3)')
ax.set_title('Gráfica 3D de x1, x2, x3')

# Mostrar la leyenda y la gráfica
ax.legend()
plt.show()
```

```python
# Normalizar los datos
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Aplicar PCA para reducir a 2 componentes principales
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Convertir el resultado a un DataFrame
data_pca_df = pd.DataFrame(data_pca, columns=['PC1', 'PC2'])

# Visualización de la varianza explicada
explained_variance = pca.explained_variance_ratio_
print("\nVarianza explicada por cada componente principal:")
print(explained_variance)

# Graficar los datos en el espacio reducido
plt.figure(figsize=(8, 6))
plt.scatter(data_pca_df['PC1'], data_pca_df['PC2'], alpha=0.7, edgecolors='k')
plt.xlabel('Componente Principal 1 (PC1)')
plt.ylabel('Componente Principal 2 (PC2)')
plt.title('Datos Transformados con PCA')
plt.grid()
plt.show()
```
