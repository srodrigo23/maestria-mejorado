# 1. Entrenados durante el entrenamiento
**Aprendizaje supervisado** Para este tipo de aprendizaje en el set de entrenamiento se alimenta el algoritmo con las soluciones deseadas, denominadas etiquetas o labels.

# Batch versus Online Learning

Otro criterio utilizado para clasificar los sistemas de aprendizaje autom√°tico es si el sistema puede aprender de forma incremental a partir de un flujo de datos entrantes.

**Batch learning** 

En el aprendizaje por lotes, el sistema no es capaz de aprender de forma incremental: debe entrenarse utilizando todos los datos disponibles. Esto generalmente requiere mucho tiempo y recursos inform√°ticos, por lo que normalmente se hace sin conexi√≥n. Primero se entrena el sistema y luego se pone en producci√≥n y funciona sin aprender m√°s; simplemente aplica lo que ha aprendido. Esto se llama aprendizaje sin conexi√≥n.

Desafortunadamente, el rendimiento de un modelo tiende a decaer lentamente con el tiempo, simplemente porque el mundo contin√∫a evolucionando mientras que el modelo permanece inalterado. Este fen√≥meno a menudo se llama deterioro del modelo o deriva de datos. La soluci√≥n es volver a entrenar peri√≥dicamente el modelo con datos actualizados.

**Online learning**
En el aprendizaje en l√≠nea, se entrena al sistema de manera incremental al alimentarlo con instancias de datos de manera secuencial, ya sea de manera individual o en grupos peque√±os llamados minilotes. Cada paso de aprendizaje es r√°pido y econ√≥mico, por lo que el sistema puede aprender sobre nuevos datos sobre la marcha, a medida que llegan.

# Principales retos en el aprendizaje autom√°tico

En escencua las dos principales tareas es seleccionar un modelo y entrenar sobre la data, en este punto solo dos cosas pueden hacerse mal **modelo err√≥neo** o **data erronea**.

**Cantidad insuficiente de datos de entrenamiento**

El aprendizaje autom√°tico a√∫n no est√° del todo desarrollado; se necesitan muchos datos para que la mayor√≠a de los algoritmos de aprendizaje autom√°tico funcionen correctamente. **Datos de entrenamiento no representativos**
Para generalizar bien, es fundamental que los datos de entrenamiento sean representativos de los nuevos casos a los que desea generalizar. Esto es as√≠ tanto si utiliza aprendizaje basado en instancias como si utiliza aprendizaje basado en modelos.

Es fundamental utilizar un conjunto de entrenamiento que sea representativo de los casos a los que se desea generalizar. Esto suele ser m√°s dif√≠cil de lo que parece: si la muestra es demasiado peque√±a, se producir√° ruido de muestreo (es decir, datos no representativos como resultado del azar), pero incluso muestras muy grandes pueden no ser representativas si el m√©todo de muestreo es defectuoso. Esto se denomina sesgo de muestreo.

**Baja calidad en la data**
Obviamente, si sus datos de entrenamiento est√°n llenos de errores, **valores at√≠picos y ruido (por ejemplo, debido a mediciones de mala calidad)**, ser√° m√°s dif√≠cil para el sistema detectar los patrones subyacentes, por lo que es menos probable que su sistema funcione bien.

A menudo, vale la pena el esfuerzo de dedicar tiempo a limpiar sus datos de entrenamiento.La mayor√≠a de los cient√≠ficos de datos dedican una parte importante de su tiempo a hacer precisamente eso.

Algunas instancias son claramente valores at√≠picos, puede ser √∫til simplemente descartarlas o intentar corregir los errores manualmente.
Si a algunas instancias les faltan algunas caracter√≠sticas (por ejemplo, el 5 % de sus clientes no especific√≥ su edad), debe decidir si desea ignorar este atributo por completo, ignorar estas instancias, completar los valores faltantes (por ejemplo, con la edad media) o entrenar un modelo con la caracter√≠stica y un modelo sin ella.

**Caracter√≠sticas irrelevantes**
El sistema solo podr√° aprender si los datos de entrenamiento contienen suficientes caracter√≠sticas relevantes y no demasiadas irrelevantes. Una parte fundamental del √©xito de un proyecto de aprendizaje autom√°tico es crear un **buen conjunto de caracter√≠sticas para entrenar**. Este proceso, llamado ingenier√≠a de caracter√≠sticas, implica los siguientes pasos:

1. Selecci√≥n de caracter√≠sticas (seleccionar las caracter√≠sticas m√°s √∫tiles para entrenar entre las caracter√≠sticas existentes)

3. Creaci√≥n de nuevas caracter√≠sticas mediante la recopilaci√≥n de nuevos datos

**Sobreajuste de los datos de entrenamiento**
En el aprendizaje autom√°tico el sobreajuste significa que el modelo funciona bien con los datos de entrenamiento, pero no se generaliza bien.


**Ajuste insuficiente de los datos de entrenamiento**
Como puede suponer, el ajuste insuficiente es lo opuesto al ajuste excesivo: ocurre cuando el modelo es demasiado simple para aprender la estructura subyacente de los datos.
Las principales opciones para solucionar este problema:

# Testeo y validacion del MODELO

La √∫nica forma de saber qu√© tan bien se generalizar√° un modelo a nuevos casos es probarlo en casos nuevos. Una forma de hacerlo es poner el modelo en producci√≥n y monitorear su rendimiento. **Una mejor opci√≥n es dividir los datos en dos conjuntos:** el conjunto de entrenamiento y el conjunto de prueba. Como estos nombres implican, se entrena el modelo utilizando el conjunto de entrenamiento y se prueba utilizando el conjunto de prueba. Sin embargo, en la pr√°ctica, los analistas de datos trabajan con tres conjuntos de ejemplos etiquetados:

1) conjunto de entrenamiento,
2) conjunto de validaci√≥n y
3) conjunto de prueba.


El conjunto de entrenamiento suele ser el m√°s grande y se utiliza para construir el modelo. Los conjuntos de validaci√≥n y prueba tienen aproximadamente el mismo tama√±o, mucho m√°s peque√±os que el tama√±o del conjunto de entrenamiento. No existe una proporci√≥n √≥ptima para dividir el conjunto de datos en estos tres subconjuntos. En el pasado, la regla general era utilizar el 70 % del conjunto de datos para entrenamiento, el 15 % para validaci√≥n y el 15 % para prueba. **Cu√°l es la raz√≥n para tener tres conjuntos y no uno?**. Un algoritmo trivial que simplemente memorice todos los ejemplos de entrenamiento y luego use la memoria para "predecir" sus salidas no cometer√° errores cuando se le pida que prediga las salidas de los dataset de entrenamiento, pero un algoritmo de este tipo ser√≠a in√∫til en la pr√°ctica. **Lo que realmente queremos es que nuestro modelo prediga bien los ejemplos que el algoritmo de aprendizaje no vio**. Por lo tanto, queremos un buen rendimiento en un conjunto de reserva.

**¬øPor qu√© necesitamos dos conjuntos de reserva y no uno?** Usamos el conjunto de validaci√≥n para:
1. Elegir el algoritmo de aprendizaje
2. Encontrar los mejores valores de los hiperpar√°metros. 
Usamos el conjunto de prueba para evaluar el modelo antes de entreg√°rselo al cliente o ponerlo en producci√≥n.

**NOTA: La regla m√°s importante a recordar es que tanto el conjunto de validaci√≥n como el conjunto de prueba deben ser lo m√°s representativos posible de los datos que espera utilizar en producci√≥n.**
