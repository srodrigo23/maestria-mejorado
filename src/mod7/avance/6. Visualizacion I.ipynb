# Visualizacion de datos

La creación de visualizaciones informativas (a veces llamadas gráficos) es una de las tareas más importantes en el análisis de datos. Puede ser parte del proceso exploratorio, por ejemplo, para ayudar a identificar valores atípicos o transformaciones de datos necesarias, o como una forma de generar ideas para modelos. 

Primeramente veremos **matplotlib** un paquete de gráficos de escritorio diseñado para crear gráficos (principalmente bidimensionales) de calidad de publicación y puede exportar visualizaciones a todos los formatos de gráficos vectoriales y rasterizados comunes (PDF, SVG, JPG, PNG, BMP, GIF, etc.)

Con el tiempo, matplotlib generó una serie de kits de herramientas complementarios para la visualización de datos que utilizan matplotlib para su representación gráfica subyacente. Uno de ellos es **seaborn**, que exploraremos más adelante.

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))

ax3 = fig.add_subplot(2,2,3)
```

```python
#Otra de declarar gráficos
fig, axes = plt.subplots(2,3) 
axes[0][0].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
```

```python
axes #array
```

Este otro metodo es muy util dado que los axes al ser un array pueden ser indexados facilmente como un vector de dos dimensiones, ej axes[0 1].
Para este metodo es posible indicar si nuestros ejes X y Y son comunes, esto es muy util cuando queremos comparar datos que estan en la misma escala. En caso de no indicar si se comparten ejes matplotlib asigna una autoescala independiente por cada subgrafico.

```python
#Es posible ajustar los interespacios en las subgraficas.
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='b', alpha=0.5)
        plt.subplots_adjust(wspace=0, hspace=0) #adicionamos este metodo
```

# Colores, estilos de linea y marcadores

La libreria nos permite controlar los colores y estilos de linea de nuestros graficos, para ello se adicionan un par de elementos dentro el argumento de la funcion.

```python
#plt.plot(np.random.randn(30).cumsum())
#plt.plot(np.random.randn(30).cumsum(), linestyle = '--', color ='g') # adicionamos dos elementos para cambiar estilo grafico
plt.plot(np.random.randn(30).cumsum(),linestyle = '--', color ='#C0CE0E') #acepta formato del color en hexadecimal
```

```python
#tambien podemos poner marcadores
plt.plot(np.random.randn(30).cumsum(),linestyle = '--', color ='#FF0E0F', marker = 'o') 
```

```python
#Tambien es posible modificar el estilo de la grafica
data = np.random.randn(30).cumsum()
plt.plot(data, 'g--', label='Por defecto')
plt.plot(data, 'k-', drawstyle='steps-post', label='estilo discreto')
plt.legend(loc ='best')
```

# Etiquetas, legendas y marcas

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())

ticks = ax.set_xticks([0, 250, 500, 750, 1000]) #añadimos marcas en el eje X
#Con la linea de abajo añadimos etiquetas a esas marcas "ticks"
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')
ax.set_title('Grafica Uno') #ponelos el título
ax.set_xlabel("Etapas")
```

```python
#Añadiendo legendas

fig = plt.figure(); 
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(100).cumsum(), 'g', label='linea uno')
ax.plot(np.random.randn(100).cumsum(), 'r--', label='linea dos')
ax.plot(np.random.randn(100).cumsum(), 'k.', label='linea 3')
ax.legend(loc='best') #busca poner la legenda en el mejor lugar posible
```
