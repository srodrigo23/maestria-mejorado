**Dimensionalidad importa:**

Los algoritmos y operaciones en NumPy y bibliotecas como pandas y scikit-learn suelen esperar datos con una dimensionalidad específica (1D o 2D).

Por ejemplo, un modelo de machine learning podría fallar si recibe un array 1D cuando espera un array 2D.

**Slicing para generalización:**

La expresión arr2D[2:, :] permite obtener subconjuntos dinámicos, lo que es útil para manejar arrays de tamaños desconocidos o al trabajar con lotes de datos.
Coherencia con otras operaciones:

Operaciones matemáticas en NumPy, como sumar filas o columnas, necesitan arrays de una dimensionalidad específica. Saber cuándo estás trabajando con un array 1D o 2D es crucial.

# Funciones Universales

Una función universal, o ufunc, es una función que realiza **operaciones elemento por elemento** sobre los datos en ndarrays. Funciones simples que toman uno o más valores escalares y producen uno o más resultados escalares.

```python
vectorX = np.arange(30,50)
vectorX
```

```python
nv = np.sqrt(vectorX)
nv
```

```python
np.round(nv,2)
```

# Métodos estadísticos Numpy

Existen un conjunto de funciones matemáticas que calculan estadísticas sobre una matriz completa o sobre los datos a lo largo de un eje en particular

```python
vec = [[1,2,3],[1,4,5],[1,6,7]] # [6,10,14] = 30/9
vec = np.array(vec)
vec.mean()
```
