# Control del flujo Clausulas If - else - elif

```python
condicion = input('Introduzca un numero de 1 al 10 ')
if int(condicion) > 5:
    print('Numero mayor a 5')
else:
    print('Numero menor a 5')
```

Realize un codigo donde se introduzca un numero, el codigo debe indicar en pantalla si: 
1. El numero es menor a 5
2. Si el numero esta en el rango de 5 al 10
3. Si el numero es mayor a 10

En caso de introducir un valor no numerico debe generar mensaje de error.

```python
condicion = input('Introduzca un numero de 1 al 20... ')

#Manejo de una excepcion, y dos condiciones sucesivas
try:
    valor = int(condicion)
    
    if valor < 5:
        print('Numero es menor a 5')
    
    elif (valor>=5 and valor<=10):
        print('Numero esta en el rango de 5 a 10')
    else:
        print("El numero es mayor a 10")
except:
    print('No se introdujo un numero, vuela a intentar')
```

```python
#Limpiando datos 
states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda','south   carolina##', 'West virginia?']

import re
#Para utilizar Expresiones Regulares, Python provee el módulo re

def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip() #Metodo str para remover espacios en blanco
        value = re.sub('[!#?]', '', value) #función para reemplazar cadenas de texto usando expresiones regulares
        value = value.title() #Metodo str para generar letra mayuscula al inicio de str
        result.append(value)
    return result
```

Las expresiones regulares son patrones utilizados para encontrar y manipular cadenas de texto de manera efectiva y precisa.

La biblioteca cuenta con una amplia variedad de funciones y meta-caracteres que permiten la creación de patrones complejos de búsqueda de cadenas de texto.

```python
clean_strings(states)
```
