```python
import pandas as pd
import numpy as np

archivo = "libro.xlsx" 
datos = pd.read_excel(archivo, header=[6], sheet_name='Potosí')
datos = datos.drop('Unnamed: 0', axis=1)
datos = datos.drop(list(range(69, 76)))
datos
```

## INICIO RESPUESTA 1 ______________________________________

```python
pd.concat( [datos.iloc[:,0] , datos.iloc[:,31:] ] , axis = 1 )
```

```python
nan_index = datos[datos.isnull().all(axis=1)].index.tolist()
dfs = []
for i in range(0,len(nan_index)-1):
    dfs.append(datos.iloc[nan_index[i]+1:nan_index[i+1],:].reset_index(drop=True))
dfs.append(datos.iloc[nan_index[-1]+1:,:].reset_index(drop=True))
dfs[0]
```

```python
result1 = pd.DataFrame(columns=['group', 'producto mas elaborado', 'cantidad'])
for i in range(0,len(dfs)):
    auxDf = dfs[i][['DESCRIPCIÓN','2022-2023(p)']].iloc[1:]
    group = dfs[i][['DESCRIPCIÓN']].iloc[0,0]
    maxIndex = auxDf['2022-2023(p)'].idxmax()
    maxProd = auxDf['2022-2023(p)'].max()
    result1.loc[i] = [group, auxDf.loc[maxIndex, 'DESCRIPCIÓN'], maxProd]
result1
```

## INICIO RESPUESTA 2

```python
promedios = pd.DataFrame(columns=['group', 'promedio de crecimiento'])
for i in range(0,len(dfs)):
    groupPromedio = dfs[i][['DESCRIPCIÓN']].iloc[0,0]
    promedio = dfs[i].iloc[[0],-10:].mean(axis=1).iloc[0]
    promedios.loc[i] = [ groupPromedio, promedio ]
promedios
    
```

```python
evolucion = dfs[0].iloc[[0]]
for i in range(1,len(dfs)):
    evolucion = pd.concat([evolucion,dfs[i].iloc[[0]]])

evolucion = evolucion.drop(columns=['promedio_producto'])
evolucion = evolucion.reset_index(drop=True)
groups_df = evolucion.iloc[:,[0]]
values_df = evolucion.iloc[0:,-10:]
evolucion_gruops = pd.concat([groups_df,values_df], axis=1)
evolucion_gruops


```

```python
import matplotlib.pyplot as plt

years = evolucion_gruops.columns[1:] 

plt.figure(figsize=(12, 6))

for i, row in evolucion_gruops.iterrows():
    plt.plot(years, row[1:], marker='o', label=row['DESCRIPCIÓN'])
    plt.text(years[-1], row.iloc[-1], row['DESCRIPCIÓN'], fontsize=10, va='center')

# Personalizar la gráfica
plt.title("Evolución de la Producción por Grupo (Subidas y Bajadas)", fontsize=14)
plt.xlabel("Años", fontsize=12)
plt.ylabel("Producción", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)

# Mostrar la gráfica sin una leyenda ya que los nombres están anotados
plt.show()
```
