# Modelos de regresión

La regresión lineal es una técnica de análisis de datos que predice el valor de datos desconocidos mediante el uso de otro valor de datos relacionado y conocido. Modela matemáticamente la variable desconocida o dependiente y la variable conocida o independiente como una ecuación lineal.

Los modelos de regresión lineal son relativamente simples y proporcionan una fórmula matemática fácil de interpretar para generar predicciones.

En el machine learning, los programas de computación denominados algoritmos analizan grandes conjuntos de datos y trabajan hacia atrás a partir de esos datos para calcular la ecuación de regresión lineal. Los científicos de datos primero entrenan el algoritmo en conjuntos de datos conocidos o etiquetados y, a continuación, utilizan el algoritmo para predecir valores desconocidos. 

**Relación lineal**
Debe existir una relación lineal entre las variables independientes y las dependientes. Para determinar esta relación, se crea una gráfica de dispersión (una colección aleatoria de valores x e y) para ver si caen a lo largo de una línea recta.

**Regresión Lineal simple**
Esta se define mediante la siguiente función: $y = \beta*x+\beta_1+e$ donde $\beta$ y $\beta_1$ son dos constantes desconocidas que representan pendiente de la regresion, y $e$ es el termino del error.

Podemos este tipo para por ejemplo: Lluvia y rendimiento de los cultivos, edad y estatura en niños, temperatura y expansión del mercurio metálico en un termómetro,etc

**Regresión lineal múltiple**
En el análisis de regresión lineal múltiple, el conjunto de datos contiene una variable dependiente y múltiples variables independientes. La función de línea de regresión lineal cambia para incluir más factores, de la siguiente manera:
$y= \beta_0 * x_0+\beta_1 * x_1+\beta_2 * x_2+... \beta_n * x_n+ e$

A medida que aumenta el número de variables predictivas, las constantes β también aumentan en consecuencia.

**La regresión lineal múltiple modela múltiples variables y su impacto en un resultado** Por ejemplo: lluvia, temperatura y uso de fertilizantes en el rendimiento de los cultivos, dieta y ejercicio sobre enfermedades cardíacas, crecimiento salarial e inflación en las tasas de préstamos hipotecarios, etc.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

tienda = pd.read_csv('Store_CA.csv')
tienda.corr(numeric_only = True)
```

```python
# Definimos area del grafico
plt.figure(figsize=(14, 7))
sns.scatterplot(data=tienda, x='StoreSize', y='MonthlySalesRevenue', color = '#443A83')


# Personalizar etiquetas y título
plt.title("Relacion entre tamaño de la tienda con los ingresos mensuales", fontsize=16)
plt.xlabel("inch", fontsize=12)
plt.ylabel("Ingresos mes USD", fontsize=12)

# Definimos area del grafico
plt.figure(figsize=(14, 7))
sns.scatterplot(data=tienda, x='ProductVariety', y='MonthlySalesRevenue', color = '#456B47')


# Personalizar etiquetas y título
plt.title("Relacion entre tamaño de la tienda con variedad de productos", fontsize=16)
plt.xlabel("", fontsize=12)
plt.ylabel("Ingresos mes USD", fontsize=12)

# Mostrar el gráfico
plt.show()
```

```python
tienda.shape
```

```python
#Primero realizaremos una regresion univariable
#Debemos separar la entrada y la salida
#Entrada
X  =  tienda['StoreSize']
#Salida
y = tienda['MonthlySalesRevenue']

from sklearn.model_selection import train_test_split
# 65% - training set (X_train, y_train)
# 35% - test set (X_test, y_test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35)
```

```python
X_train
```

```python
#Hasta aca X_train & X_test son instancias de Pandas (Series)
#Debemos convertirlas 

X_train = X_train.to_frame()
X_test = X_test.to_frame()
```

```python
type(X_train)
```

```python
from sklearn.linear_model import LinearRegression
# Creamos la instancia para el modelo lineal

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#Ya tenemos el modelo, para evaluarlo necesitamos calcular metricas MAE, MSE y RMSE
predicted = model.predict(X_test)

#Cremos dataset que contenga nuestros valores resultantes vs los reales
actual_vs_predicted = pd.DataFrame(
    {'Actual': y_test, 
     'Predicted':predicted}
)
```

```python
# Mostramos las primeras 5 filas
actual_vs_predicted.head()
```

```python
# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))
plt.plot(actual_vs_predicted.index, actual_vs_predicted['Actual'], label='Valores reales', marker='o')
plt.plot(actual_vs_predicted.index, actual_vs_predicted['Predicted'], label='Valores predecidos', marker='s')

# Añadir etiquetas y leyenda
plt.title('Gráfico de Líneas Combinadas', fontsize=14)
plt.xlabel('Índice', fontsize=12)
plt.ylabel('Ganacia mensual', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
```

```python
actual_vs_predicted.shape
```

```python
actual_vs_predicted = actual_vs_predicted.reset_index(drop = True)
actual_vs_predicted
```

```python
# Crear el gráfico de líneas
plt.figure(figsize=(28, 6))
plt.plot(actual_vs_predicted.index, actual_vs_predicted['Actual'], label='Valores reales', marker='o')
plt.plot(actual_vs_predicted.index, actual_vs_predicted['Predicted'], label='Valores predecidos', marker='s')

# Añadir etiquetas y leyenda
plt.title('Gráfico valores reales vs valores predecidos', fontsize=14)
plt.xlabel('Índice', fontsize=12)
plt.ylabel('Ganacia mensual', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
```

```python
#Cual es el error numericamente hablando???

from sklearn.metrics import (
    mean_absolute_error, # MAE
    mean_squared_error # MSE 
                             )
    
MAE = mean_absolute_error( y_true=y_test, # valores actuales
                           y_pred=predicted # valores predecidos
                         )
MAE.round(2)
```

```python
MSE = mean_squared_error(
    y_true=y_test, # actual values
    y_pred=predicted # predicted values
)
MSE.round(2)
```

```python
# Raiz cuadrada de MSE nos da el RMSE
RMSE = MSE**(1/2)
RMSE.round(2)
```
